Import all the necessary library modules

```python
from gtts import gTTS
import pytesseract as tess
from langdetect import detect
import urllib3
import cv2
from IPython.display import Image
import shutil
from IPython.display import Audio
urllib3.disable_warnings()
```

Display the image

```python
image_path = '/content/drive/MyDrive/Dataset/Images/en.jpg'
Image(filename=image_path)
```

```python
from PIL import Image

img = Image.open(image_path)
text1 = tess.image_to_string(img)
language = detect(text1)
print(language)
```

We assume the text language is either English or Bengali. So we determine the language and do the detection.

```python
if language=="en" :
  print(text1)
  audio = gTTS(text=text1, lang=language, slow=False)
  audio.save("/content/1.wav")
  #audio.save("1.wav")
else:
  text1 = tess.image_to_string(img ,lang='ben')
  language = detect(text1)
  print(text1)
  audio = gTTS(text=text1, lang=language, slow=False)
  audio.save("/content/1.wav")
  #audio.save("1.wav")
```
